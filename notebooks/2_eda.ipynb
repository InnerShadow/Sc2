```python
from dateutil.relativedelta import relativedelta

df = df.sort_values(['player', 'date']).reset_index(drop = True)
df
```

```python
from tqdm import tqdm

def add_winrate_features(player_df):
    player_df = player_df.copy()

    for m in range(1, 13):
        for race in ['P', 'T', 'Z', 'all']:
            col_match = f'wr_match_{race}_{m}m'
            col_game = f'wr_game_{race}_{m}m'

            values_match = []
            value_games = []

            for idx, row in player_df.iterrows():
                end = row['date']
                start = end - relativedelta(months = m)

                if race != 'all':
                    window = player_df[
                        (player_df['date'] < end) &
                        (player_df['date'] >= start) &
                        (player_df['opponent_race'] == race)
                    ]
                else:
                    window = player_df[
                        (player_df['date'] < end) &
                        (player_df['date'] >= start)
                    ]
                # end if

                values_match.append(window['player_win_prc'].mean())
                value_games.append(window['target'].mean())
            # end for

            player_df[col_match] = values_match
            player_df[col_game] = value_games
        # end for
    # end for

    return player_df
# end def

df = (
    df
    .groupby('player', group_keys = False)
    .apply(add_winrate_features)
)
```

```python
from tqdm import tqdm

def add_winrate_history_features(player_df):
    player_df = player_df.copy()

    for m in range(1, 13):
        col_match = f'wr_match_history_{m}m'
        col_game = f'wr_game_history_{m}m'

        values_match = []
        value_games = []

        for idx, row in player_df.iterrows():
            oponent = row['opponent']
            end = row['date']
            start = end - relativedelta(months = m)

            window = player_df[
                (player_df['date'] < end) &
                (player_df['date'] >= start) &
                (player_df['opponent'] == oponent)
            ]

            values_match.append(window['player_win_prc'].mean())
            value_games.append(window['target'].mean())
        # end for

        player_df[col_match] = values_match
        player_df[col_game] = value_games
    # end for

    return player_df
# end def

df = (
    df
    .groupby('player', group_keys = False)
    .apply(add_winrate_history_features)
)
```

```python
import numpy as np

periods = [f"{i}m" for i in range(1, 13)]
race_mapping = {}

for period in periods:
    for race in ['P', 'T', 'Z']:
        race_mapping[(period, 'match', race)] = f'wr_match_{race}_{period}'
        race_mapping[(period, 'game', race)] = f'wr_game_{race}_{period}'
    # end for
# end for

def get_wr_by_race(row, period, stat_type='match'):
    opponent_race = row['opponent_race']
    
    if opponent_race == 'R':
        return np.nan
    # end if
    
    col_name = race_mapping.get((period, stat_type, opponent_race))
    
    if col_name and col_name in row:
        return row[col_name]
    # end if
    return np.nan
# end def

for period in periods:
    df[f'wr_match_vs_opponent_{period}'] = df.apply(
        lambda x: get_wr_by_race(x, period, 'match'), axis=1
    )
    
    df[f'wr_game_vs_opponent_{period}'] = df.apply(
        lambda x: get_wr_by_race(x, period, 'game'), axis=1
    )
# end for

for period in periods:
    conditions = [
        df['opponent_race'] == 'P',
        df['opponent_race'] == 'T',
        df['opponent_race'] == 'Z'
    ]
    
    choices_match = [
        df[f'wr_match_P_{period}'],
        df[f'wr_match_T_{period}'],
        df[f'wr_match_Z_{period}']
    ]
    
    choices_game = [
        df[f'wr_game_P_{period}'],
        df[f'wr_game_T_{period}'],
        df[f'wr_game_Z_{period}']
    ]
    
    df[f'wr_match_vs_opponent_race_{period}'] = np.select(
        conditions, choices_match, default=np.nan
    )
    df[f'wr_game_vs_opponent_race_{period}'] = np.select(
        conditions, choices_game, default=np.nan
    )
# end for

df
```

```python
df[['player', 'player_race', 'opponent', 'opponent_race', 'tier']] = df[['player', 'player_race', 'opponent', 'opponent_race', 'tier']].astype('category')
```

```python
from xgboost import XGBClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import roc_auc_score

X_train, X_test, y_train, y_test = train_test_split(
    df.drop(columns = ['player_score', 'opponent_score', 'player_win_prc', 'target', 'date']),
    df['target'],
    test_size = 0.2
)

model = XGBClassifier(n_jobs = -1, enable_categorical = True)
model.fit(X = X_train, y = y_train)
roc_auc_score(y_test, model.predict_proba(X_test)[:, 1])
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test, model.predict(X_test)))
```

```python
print(classification_report(y_train, model.predict(X_train)))
```
