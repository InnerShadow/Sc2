```python
%cd './..'
import pandas as pd
```

```python
df = pd.read_csv('./data/dataset.csv', sep = ';')
df
```

```python
df['date'] = pd.to_datetime(df['date'])
df
```

```python
df = df.drop(columns = ['player_id', 'opponent_id', 'score_match', 'url', 'tournament'])
df
```

```python
print(df.shape)
df = df[df['player_score'] != df['opponent_score']].reset_index(drop = True)
print(df.shape)
```

```python
add_rows = []

for it, row in df.iterrows():
    add_rows.append({
        'date' : row['date'],
        'player' : row['opponent'],
        'player_race' : row['opponent_race'],
        'player_score' : row['opponent_score'],
        'opponent' : row['player'],
        'opponent_race' : row['player_race'],
        'opponent_score' : row['player_score'],
        'tier' : row['tier']
    })
# end for

df_add = pd.DataFrame(add_rows)
df = pd.concat([df, df_add]).reset_index(drop = True)
df
```

```python
df['best_of'] = 2 * df[['player_score', 'opponent_score']].max(axis = 1) - 1
df
```

```python
df['player_win_prc'] = (df['player_score'] / (df['player_score'] + df['opponent_score'])).round(4)
df
```

```python
df['target'] = (df['player_score'] > df['opponent_score']).astype(int)
df
```

```python
from dateutil.relativedelta import relativedelta

df = df.sort_values(['player', 'date']).reset_index(drop = True)
df
```

```python
from tqdm import tqdm

def add_winrate_features(player_df):
    player_df = player_df.copy()

    for m in range(1, 13):
        for race in ['P', 'T', 'Z']:
            col_match = f'wr_match_{race}_{m}m'
            col_game = f'wr_game_{race}_{m}m'

            values_match = []
            value_games = []

            for idx, row in player_df.iterrows():
                end = row['date']
                start = end - relativedelta(months = m)

                window = player_df[
                    (player_df['date'] < end) &
                    (player_df['date'] >= start) &
                    (player_df['opponent_race'] == race)
                ]

                values_match.append(window['player_win_prc'].mean())
                value_games.append(window['target'].mean())
            # end for

            player_df[col_match] = values_match
            player_df[col_game] = value_games
        # end for
    # end for

    return player_df
# end def

df = (
    df
    .groupby('player', group_keys = False)
    .apply(add_winrate_features)
)
```

```python
from xgboost import XGBClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import roc_auc_score

X_train, X_test, y_train, y_test = train_test_split(
    df.drop(columns = ['player_score', 'opponent_score', 'player_win_prc', 'target', 'date']),
    df['target'],
    test_size = 0.2
)

model = XGBClassifier(n_jobs = -1)
model.fit(X = X_train, y = y_train)
roc_auc_score(y_test, model.predict_proba(X_test)[:, 1])
```
